## Whos the Greatest Basketball Player of All time? ##

In our household, we continue to debate whos the greatest, MJ, Lebron, Kobe, Bill Russell, etc
I decided to use data to informe the decision. 

Primary data source - https://www.nba.com/stats/alltime-leaders?StatCategory=PTS&SeasonType=Regular+Season&PerMode=PerGame

### Create a CSV file with key metrics for each player ###
I pulled all time, and playoff metrics, and saved into a file goat.csv

```python
# Import csv file intoa pandas dataframe to conduct further analysis
import pandas as pd

data = pd.read_csv('goat.csv')
df = pd.DataFrame(data)
df
```

### Create difference columns, comparing players in each Row. ### 
MJ is top of list, and Lebron 2, so Lebrons data will be compared to MJ, Then Kobe to Lebron, etc.

```python
#diff=df['Games'].diff()
df['game-diff']=df['Games'].diff()
df['season-diff']=df['Seasons'].diff()
df['pts-diff']=df['PTS-all-time'].diff()
df['ppg-diff']=df['Career-Pt-Avg'].diff()
df['mvp-diff']=df['MVP'].diff()
df['playoff-ppg-diff']=df['PlayoffPts-average'].diff()
df['allstar-mvp-diff']=df['Allstar-mvp'].diff()
df['DPOY-diff']=df['DPOY'].diff()
df['adt-diff']=df['ADT'].diff()

df
```

```python
#Convert results to csv file, and save to local drive
df.to_csv('goat-data-cleaned.csv')
```
